````r
options(repr.plot.width = 8, repr.plot.height = 3)
p = ggplot(penguins_long, aes(x = species, y = value, color = sex))
p = p + geom_point(size = .02, alpha = .05) + facet_wrap(vars(name), nrow = 1, scales = "free_y")
p = p + stat_summary(fun = median, 
                     fun.min = median, 
                     fun.max = median, 
                     geom = "crossbar", 
                     width = 0.75, 
                     size = 0.2,
                     mapping = aes(group = species))
p = p + theme(axis.text.x = element_text(hjust = 1, angle = 45))
p = p + geom_jitter(position = position_jitter(width = .3, height = 0))
p
````
